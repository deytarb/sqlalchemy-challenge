### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months

df['month'] = pd.DatetimeIndex(df['date']).month
jun = df.loc[df['month'] == 6]
dec = df.loc[df['month'] == 12]
df = df.drop(['month'], axis=1)
jun = jun.drop(['month'], axis=1)
dec = dec.drop(['month'], axis=1)
print(jun.head())
print(dec.head())
```

```python
# Identify the average temperature for June

aveJun = jun['tobs'].mean()

print(aveJun)
```

```python
# Identify the average temperature for December

aveDec = dec['tobs'].mean()

print(aveDec)
```

```python
# Create collections of temperature data

collectionJun = jun['tobs'].to_dict()
collectionDec = jun['tobs'].to_dict()

print(collectionJun)
print(collectionDec)
```

```python
# Run paired t-test

stats.ttest_ind(jun['tobs'], dec['tobs'])
```
